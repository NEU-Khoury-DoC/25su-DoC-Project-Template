```python
import pandas as pd
import os

# Step 1: Read the CSV file
df = pd.read_csv('../datasets/MEGAFRAME_CLEANEDV2.csv')
```

```python
# Step 2: Select 5 random rows
df_sample = df.sample(n=5, random_state=42)  # random_state makes it reproducible
df_sample
```

```python
# Step 3: Define the nested folder path and file name
folder_path = 'data/samples'  # example nested path
file_name = 'features_and_target_sample_dataset.csv'
full_path = os.path.join(folder_path, file_name)

# Step 4: Create the nested folder if it doesn't exist
os.makedirs(folder_path, exist_ok=True)

# Step 5: Save the sample CSV
df_sample.to_csv(full_path, index=False)
```
